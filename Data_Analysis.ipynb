```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate total sales by Representative and Item
rep_item_sales = df.groupby(['Rep', 'Item'])['Total'].sum().reset_index()

# Create a horizontal bar chart
plt.figure(figsize=(14, 8))
sns.barplot(x='Total', y='Rep', hue='Item', data=rep_item_sales)
plt.title('Total Sales by Representative and Item')
plt.xlabel('Total Sales')
plt.ylabel('Representative')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

region_sales = df.groupby('Region')['Total'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Total', data=region_sales)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df['OrderDate'] = pd.to_datetime(df['OrderDate'])

sales_by_date = df.groupby('OrderDate')['Total'].sum().reset_index()

plt.figure(figsize=(12, 7))
sns.lineplot(x='OrderDate', y='Total', data=sales_by_date, marker='o')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

item_sales = df.groupby('Item')['Total'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Item', y='Total', data=item_sales)
plt.title('Total Sales by Item')
plt.xlabel('Item')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
