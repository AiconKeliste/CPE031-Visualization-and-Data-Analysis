```python
#Creating DataFrame
import pandas as pd

data = {
    'apples':[3,2,0,1],
    'oranges':[0,3,7,2]
}

purchases = pd.DataFrame(data)

print(purchases)
```

```python
#DataFrame Index
import pandas as pd

data = {
    'apples':[3,2,0,1],
    'oranges':[0,3,7,2]
}

purchases = pd.DataFrame(data, index = ['June', 'Robert', 'Lily', 'David'])

print(purchases)
```

```python
#Locate
import pandas as pd

data = {
    'apples':[3,2,0,1],
    'oranges':[0,3,7,2]
}

purchases = pd.DataFrame(data, index = ['June', 'Robert', 'Lily', 'David'])
purchases.loc['June']
```

```python
#Importing DataFrame
import pandas as pd
data = {
    'apples':[3,2,0,1],
    'oranges':[0,3,7,2]
}

purchases = pd.DataFrame(data, index = ['June', 'Robert', 'Lily', 'David'])

purchases.to_excel('ComputerData.xlsx')#Convert to xlsx
```
