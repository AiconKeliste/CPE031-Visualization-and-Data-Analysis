**4. Procedures:**

Step 1. Import Libraries and Create Sample Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)
df
```

Step 2. Know Your Audience — Compare Chart Styles

```python
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="steelblue")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()
```

**Task 1:** Create one additional visualization suited for a different audience (e.g., a customer, investor, or teammate).

```python
#Code Here
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="red")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()

# Investor audience: prefer overall contribution (pie chart)
plt.figure(figsize=(8, 8))
plt.pie(df['Sales'], labels=df['Month'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('coolwarm_r'))
plt.title('Investor View: Monthly Sales Contribution', fontsize=13)
plt.show()
```

Step 3. Know Your Message — Highlight What Matters

```python
# Highlight the highest sales month (preattentive color cue)
highlight = df["Sales"].idxmax()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Best Month in Sales", fontsize=13)
plt.show()
```

**Task 2:** Use color or annotations to emphasize a key insight (e.g., lowest point, sudden increase).

```python
#Code Here
# Highlight the lowest sales month (preattentive color cue)
highlight_lowest = df["Sales"].idxmin()

colors_lowest = ["gray" if i != highlight_lowest else "turquoise" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors_lowest)
plt.title("Highlighting the Lowest Month in Sales", fontsize=13)

# Add annotation for the lowest point
lowest_month = df.loc[highlight_lowest, "Month"]
lowest_sales = df.loc[highlight_lowest, "Sales"]
plt.annotate(f'Lowest Sales\n({lowest_sales})',
             xy=(highlight_lowest, lowest_sales),
             xytext=(highlight_lowest, lowest_sales + 50),
             ha='center',
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

Step 4. Storytelling with Charts — Beginning, Middle, End

```python
# Beginning: Context
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="skyblue")
plt.title("Beginning: Sales Started Steady", fontsize=13)
plt.show()

# Middle: Change or Challenge
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, color="orange", marker="o")
plt.title("Middle: A Drop in March Due to Low Marketing", fontsize=13)
plt.show()

# End: Resolution
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="limegreen")
plt.title("End: Recovery and Growth After March", fontsize=13)
plt.show()
```

**Task 3:** Write one or two lines describing your data story from beginning → middle → end.

*Answer here <br>
The sales started steady, but experienced a drop in March. This drop was followed by a recovery and growth in the subsequent months.

Step 5. Bringing It All Together — Presentation Design

```python
# Simple combined story chart
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", linewidth=2, color="royalblue", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", linestyle="--", color="orange", label="Marketing Spend")

plt.fill_between(df["Month"], df["Sales"], alpha=0.1, color="royalblue")
plt.title("From Effort to Impact: Marketing Spend Drives Sales Growth", fontsize=13)
plt.legend()
plt.show()
```

```python
df  = pd.read_csv("/content/drive/MyDrive/VDA/Covid19 India (Jan 20 - Mar 20).csv")
df
```

```python
# Group by State/UnionTerritory and sum the confirmed cases for Indian nationals
state_data = df.groupby('State/UnionTerritory')['ConfirmedIndianNational'].sum().reset_index()

#highest
highlight_state = state_data['ConfirmedIndianNational'].idxmax()

#different color to highlight
colors = ["skyblue" if i != highlight_state else "red" for i in range(len(state_data))]

#bar chart
plt.figure(figsize=(14, 13))
plt.bar(state_data['State/UnionTerritory'], state_data['ConfirmedIndianNational'], color=colors)

#labels
plt.title("Confirmed COVID-19 Cases in India (Jan-Mar 2020) by State/Union Territory", fontsize=12)
plt.xlabel("State/Union Territory", fontsize=12)
plt.ylabel("Number of Confirmed Indian Nationals", fontsize=12)
plt.xticks(rotation=75) # Rotate x-axis labels for better readability

#annotations
highest_state_name = state_data.loc[highlight_state, 'State/UnionTerritory']
highest_cases = state_data.loc[highlight_state, 'ConfirmedIndianNational']
plt.annotate(f'{highest_state_name}\n({highest_cases} cases)',
             xy=(highlight_state, highest_cases),
             xytext=(highlight_state, highest_cases + 50),
             ha='center',
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.tight_layout()
plt.show()
```
